# Data Download -- Online Retail II Dataset

Downloads the Online Retail II dataset from the UCI Machine Learning Repository and converts it to CSV for faster loading.


```python
import os
import urllib.request
import zipfile
import pandas as pd

DATA_DIR = os.path.join("..", "data", "raw")
os.makedirs(DATA_DIR, exist_ok=True)
```

```python
url = "https://archive.ics.uci.edu/static/public/502/online+retail+ii.zip"
zip_path = os.path.join(DATA_DIR, "online_retail_ii.zip")
xlsx_path = os.path.join(DATA_DIR, "online_retail_II.xlsx")

if os.path.exists(xlsx_path):
    print(f"Dataset already exists at {xlsx_path}")
else:
    print("Downloading Online Retail II dataset from UCI...")
    urllib.request.urlretrieve(url, zip_path)
    print(f"Downloaded to {zip_path}")

    print("Extracting...")
    with zipfile.ZipFile(zip_path, "r") as z:
        z.extractall(DATA_DIR)
    os.remove(zip_path)
    print(f"Extracted to {DATA_DIR}")
```

### Convert to CSV

```python
csv_path = os.path.join(DATA_DIR, "online_retail_II.csv")

if os.path.exists(csv_path):
    print(f"CSV already exists at {csv_path}")
else:
    print("Converting Excel to CSV...")
    df = pd.read_excel(xlsx_path, engine="openpyxl")
    df.to_csv(csv_path, index=False)
    print(f"Saved CSV to {csv_path}")
```
